# **Overview of Iris Flower Dataset**
---------------

The Iris dataset is a classic dataset in the field of machine learning and statistics.

It was introduced by the British statistician and biologist Ronald Fisher in his 1936 paper "The use of multiple measurements in taxonomic problems" as an example of linear discriminant analysis.

The dataset consists of 150 samples of iris flowers, each belonging to one of three species: **Setosa, Versicolor,** and **Virginica.**

------------------------------------------
## Columns : -
- **Sepal Length (cm):**The length of the sepal, which is the outermost part of the flower that protects the petals.
- **Sepal Width (cm):** The width of the sepal, measured at its widest point.
- **Petal Length (cm):** The length of the petal, which is the colorful inner part of the flower.
- **Petal Width (cm):** The width of the petal, measured at its widest point.
- **Species:** The species of iris flower, which can be one of three classes: Setosa, Versicolor, or Virginica.


These columns provide measurements of various parts of the iris flower, along with the corresponding species labels, making it a versatile dataset for analysis and classification tasks.

# **Iris Flower Clustering using KMeans**

------------------
Our project applied KMeans Clustering to group iris flowers into distinct clusters based

on their similarities in measurements like **sepal length, sepal width, petal length**, and **petal width**.

KMeans, a centroid-based algorithm, partitions the dataset into **K** clusters

by iteratively assigning data points to the nearest cluster centroid and updating centroids until convergence.

----------

#### Since we have already imported and read the dataset, explored its features, and visualized the data in the preceding Logistic Classification section, we can now proceed directly to feature standardization.
------------
## **Feature Standardization**

-----------

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
# scale the data
X_scaled  = scaler.fit_transform(X)
```

## **KMeans Clutering**
-----

```python
from sklearn.cluster import KMeans


k = 3  # Optimum number of clusters (as we know their 3 species of iris flowers)
kmeans = KMeans(n_clusters=k, random_state=42)
kmeans.fit(X_scaled)


# Get cluster centers and labels
centroids = kmeans.cluster_centers_
labels = kmeans.labels_
```

## **Plot the Clusters**
------

```python
# Plot the clusters
plt.figure(figsize=(8, 6))

# Visualize the clusters
colors = ['red', 'green', 'blue']  # Color mapping for clusters
labels = kmeans.labels_

# Plot the scatter plot
plt.figure(figsize=(8, 6))
for i in range(k):
    cluster_points = X_scaled[labels == i]
    plt.scatter(cluster_points[:, 0], cluster_points[:, 1], c=colors[i], label=f'Cluster {i+1}')
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], marker='x', color='black', label='Centroids')
sns.set_style('dark')
plt.xlabel('Sepal Length')
plt.ylabel('Sepal Width')
plt.title('K-means Clustering on Iris Dataset')
plt.legend()
plt.show();
```

## **Evaluate the Model**
---

```python
from sklearn.metrics import silhouette_score
# Evaluate the clustering using silhouette score
silhouette_avg = silhouette_score(iris, kmeans.labels_)
print("Silhouette Score:", silhouette_avg)

# Get the inertia (within-cluster sum of squares)
inertia = kmeans.inertia_
print("Inertia:", inertia)
```

--------------
**Obs:-**

**1** . The silhouette score of approximately **0.495** suggests that the clusters are reasonably well-separated ,indicating a good clustering solution.

**2** . he fact that both the silhouette score and inertia are provided here and the inertia value remains unchanged at **139.82** indicates that while the clusters are relatively compact.

**3** . There may be some overlapping or less clear boundaries between clusters, which is reflected in the silhouette score being slightly below **0.5**.
